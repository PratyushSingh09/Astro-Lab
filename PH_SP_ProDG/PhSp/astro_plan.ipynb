```python
from astroplan import Observer
import astropy.units as u
subaru = Observer.at_site('subaru')
iaat = Observer(latitude=48.534*u.deg,longitude=9.07*u.deg)
```

```python
from astropy.coordinates import SkyCoord
from astroplan import FixedTarget

altair = FixedTarget.from_name('Altair')
vega = FixedTarget.from_name('Vega')

M11 = FixedTarget.from_name('M11')
M34 = FixedTarget.from_name('M34')
M36 = FixedTarget.from_name('M36')
M39 = FixedTarget.from_name('M39')
lac_10 = FixedTarget.from_name('10 Lac')
aCam = FixedTarget.from_name('alpha Cam')
aCas = FixedTarget.from_name('alpha Cas')
bCas = FixedTarget.from_name('beta Cas')
gCas = FixedTarget.from_name('gamma Cas')
bDra = FixedTarget.from_name('beta Dra')
aLyr = FixedTarget.from_name('alpha Lyr')
nUma = FixedTarget.from_name('eta Uma')
```

```python
from astropy.time import Time

time = Time('2025-03-19 20:00:00')
```

```python
coordinates = SkyCoord('20h41m25.9s', '+45d16m49.3s', frame='icrs')
deneb = FixedTarget(name='Deneb', coord=coordinates)
```

```python
# iaat.target_is_up(time, altair)

# iaat.target_is_up(time, vega)

# iaat.target_is_up(time, deneb)
iaat.target_is_up(time, M11)
iaat.target_is_up(time, M34)
iaat.target_is_up(time, M36)
iaat.target_is_up(time, M39)
iaat.target_is_up(time, lac_10)
iaat.target_is_up(time, aCam)
iaat.target_is_up(time, aCas)
iaat.target_is_up(time, bCas)
iaat.target_is_up(time, gCas)
iaat.target_is_up(time, bDra)
iaat.target_is_up(time, aLyr)
iaat.target_is_up(time, nUma)
```
